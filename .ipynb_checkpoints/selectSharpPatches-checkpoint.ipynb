```python
import os
import shutil
```

```python
file = open("sharpPatchesIndices.txt","r")
indices = map(lambda x: x[1] ,enumerate(file))

destinationDir = "sharpPatches"


for i in indices :
    filePath = "patch" + str(i).rstrip() + ".png"

    print(filePath)
    if not os.path.isfile(filePath):
        print("file does not exist")
    shutil.copy2(filePath,destinationDir)
```
