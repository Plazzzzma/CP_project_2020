```python
import numpy as np
import imageio
import matplotlib.pyplot as plt
import cv2
import imgaug.augmenters as iaa
from scipy import signal
import os, os.path

#Helpers
import data_loader as dl
import kernel_estimate as ke
```

```python
#load original images

image1Path = "Data/focused_images/cropped_images/stack1_cropped.jpg"
image2Path = "Data/focused_images/cropped_images/stack2_cropped.jpg"
image3Path = "Data/focused_images/cropped_images/stack3_cropped.jpg"
image4Path = "Data/focused_images/cropped_images/stack4_cropped.jpg"
blurred_images_folder = "Data/focused_images/cropped_images/blurred_images/"
ground_truth_folder = "Data/focused_images/cropped_images/ground_truth/"

image1 = imageio.imread(image1Path)
image2 = imageio.imread(image2Path)
image3 = imageio.imread(image3Path)
image4 = imageio.imread(image4Path)

images = [image1, image2, image3, image4]

flipHor = iaa.Fliplr(1.0)
flipVer = iaa.Flipud(1.0)

rotate90 = iaa.Rot90((1),keep_size = False)
rotate180 = iaa.Rot90((2),keep_size = False)
rotate270 = iaa.Rot90((3),keep_size = False)


rotations = [rotate90, rotate180, rotate270]
flips = [flipHor, flipVer]
im_index = 0
aug_images = []
for im in images:

    imageio.imwrite(ground_truth_folder + 'gt_image'+ str(im_index) +'.png',im)
    im_index = im_index +1
    
    for rot in rotations:
        
        rot_image = rot(image = im)
        imageio.imwrite(ground_truth_folder + 'gt_image'+ str(im_index) +'.png',rot_image)
        im_index = im_index +1
        
        for flip in flips:
            
            flip_image = flip(image = rot_image)
            imageio.imwrite(ground_truth_folder + 'gt_image'+ str(im_index) +'.png',flip_image)
            im_index = im_index +1
            
        
    

            
#flipped_image = blur1(image = images[0])
#plt.imshow(flip1(images =image1))
#plt.imshow(flipped_image)
```

```python
#load blur kernels and ground truth images
ground_truth_folder = "Data/focused_images/cropped_images/ground_truth/"
selected_kernels = np.load('selected_kernels2.npy',allow_pickle=True)
#gt_images = dl.load_set(ground_truth_folder, 40)
images_names = os.listdir(ground_truth_folder)
gt_images = []
for name in images_names:
    gt_images.append(cv2.imread(ground_truth_folder + name,cv2.IMREAD_GRAYSCALE))
    
```

```python
#blur augmented data

ground_truth_final_folder = "Data/focused_images/cropped_images/ground_truth_final2/"
blurred_final_folder = "Data/focused_images/cropped_images/blurred_final2/"


blurs = selected_kernels
#for i in range(50):
#    blur = iaa.GaussianBlur(sigma = (i/(2.) + 1))
#    blurs.append(blur)
    

im_index = 400
for bl in blurs:
    for im in gt_images: 
        #blurred_image = bl(image = im)
        blurred_image = signal.convolve2d(im,bl,mode = 'same',boundary = 'wrap')
        cv2.imwrite(blurred_final_folder + 'blurred_image'+ str(im_index) +'.png',blurred_image)
        cv2.imwrite(ground_truth_final_folder + 'gt_image'+ str(im_index) +'.png',im)
        im_index = im_index +1
        print('patch' + str(im_index) + 'over ' + str(len(gt_images)*len(blurs)))
    
    
```
